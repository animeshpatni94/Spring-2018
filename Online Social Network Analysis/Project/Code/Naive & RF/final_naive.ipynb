```python
def replaceTwoOrMore(s):
    #look for 2 or more repetitions of character and replace with the character itself
    pattern = re.compile(r"(.)\1{1,}", re.DOTALL)
    return pattern.sub(r"\1\1", s)
```

```python
stop=[]
def removestopwords(filename):
    file = open(filename,'rt')
    for i in file:
        i = i.strip()
        stop.append(i)
    file.close()
    return stop
```

```python
def featurize(tweet):
    mat=[]
    w=tweet.split()
    for i in w:
        i = replaceTwoOrMore(i)#replace the words that occur two or more times
        i = i.strip('\'"?,.')#remove the basic punctuatuions, could have used string.punctuatuion instead
        val = re.search(r"^[a-zA-Z][a-zA-Z0-9]*$", i)#checking if any words starts with in an alpha
        if(val is None or i in stop):
            continue
        else:
            mat.append(i.lower())
    #print(mat)
    return mat
```

```python
#read tweets one by one and pre process it to the best way I can
#st = open('StopWords.txt','rt')
#stop = removestopwords('StopWords.txt')
tweets = csv.reader(open('ManuallyTag_naive.csv','rt'), delimiter=',', quotechar='|')
t=[]
l=[]
#getting tokens
for i in tweets:
    senti = i[7]
    tweet = i[3]
    #processed = preprocess(tweet)
    f = featurize(tweet)
    l.extend(f)
    t.append((f,senti))
l = list(set(l)) #done with the duplicates
#now comes in NLTK to vectorize it in one line
#wish could use this in Assignment 2 :P
train = nltk.classify.util.apply_features(extract_features,t)
```

```python
#start extract_features
#this method will tell us whether we have feautres present in the incoming new tweet. will give us 0 or 1 or we can also have true or false
def extract_features(tweet):
    tweet_words = set(tweet)
    features = {}
    for word in l:
        if word in tweet_words:
            features['there is:(%s)' % word] = True
        else:
            features['there is:(%s)' % word] = False
    return features
```

```python
NBClassifier = nltk.NaiveBayesClassifier.train(train)
testTweet = 'Bad bad world'
#processedTestTweet = preprocess(testTweet)
print (NBClassifier.classify(extract_features(featurize(testTweet))))
```
